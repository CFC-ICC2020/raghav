# Salary Slab Vs Previous Contract Status

```python
total_applications = (updated_curr_app[['COUNT_APPROVED', 'COUNT_REFUSED' , 'COUNT_CANCELED', 'COUNT_UNUSED', 'SALARY_RANGE']].groupby(by = 'SALARY_RANGE').sum()).sum(axis = 1).reset_index()
total_applications['Count'] = total_applications[0]
total_applications.drop(0, axis = 1, inplace = True)
```

```python
sal_status_count = (updated_curr_app[['SALARY_RANGE', 'COUNT_APPROVED', 'COUNT_REFUSED', 'COUNT_CANCELED', 'COUNT_UNUSED' ]].groupby(by = 'SALARY_RANGE').sum()).reset_index()
sal_status_count
```

```python
sal_sta_total = pd.merge(left = sal_status_count, right = total_applications, on = 'SALARY_RANGE', how = 'inner')
sal_sta_total
```

```python
# Calculating Percentage of applicants under diff slabs

## Here we calculated percents corresponding to Salary Slabs 

### i.e, Total Approved (or) Total Refused (or) Total Unused (or) Total Canceled / Total Applications under that salary slab

sal_sta_total['PERCENT_APPROVED'] = sal_sta_total.COUNT_APPROVED/sal_sta_total.Count * 100
sal_sta_total['PERCENT_REFUSED'] = sal_sta_total.COUNT_REFUSED/sal_sta_total.Count * 100 
sal_sta_total['PERCENT_CANCELED'] = sal_sta_total.COUNT_CANCELED/sal_sta_total.Count * 100
sal_sta_total['PERCENT_UNUSED'] = sal_sta_total.COUNT_UNUSED/sal_sta_total.Count * 100
sal_sta_total.dropna(inplace = True)
```

```python
# Bar Plots showing different Percents Vs Salary Slabs

plt1 = sal_sta_total[['SALARY_RANGE', 'PERCENT_APPROVED', 'PERCENT_REFUSED', 'PERCENT_CANCELED', 'PERCENT_UNUSED']].set_index('SALARY_RANGE').plot.bar(figsize = (15,7))
for bar in plt1.patches:
    plt1.annotate("{}%".format(format(bar.get_height(), '.2f')), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height() + 3 ), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points',
                   weight = 'bold', fontsize=10,rotation = 90)

plt1.set_xlabel("Percentage of Applications", fontweight = 'bold', fontsize =12)
plt1.set_ylabel('Salary Ranges', fontweight = 'bold', fontsize =12)
plt1.legend(bbox_to_anchor = (1.0,1.0))
plt.savefig('Salary Ranges Vs Percent of status of applications', bbox_inches = 'tight')
plt.show()
```

`Inference`: We can see here that Slab_8 has highest approval percentage of loans.

Approved status of loans follows below pattern regarding Salary Ranges

`Slab 8 > Slab 4 > Slab 1 > Slab 6 > Slab 2 > Slab 3 > Slab 5`

## Ratio of Client types

```python
# Bar plot showing different client types


fig, ax = plt.subplots(1,2)
fig.tight_layout(pad = 3.0)

plot = (updated_curr_app.NAME_CLIENT_TYPE.value_counts(normalize = True)*100).plot.bar(ax = ax[0], figsize = (15,7))
for bar in plot.patches:
        plot.annotate("{}%".format(format(bar.get_height(), '.2f')), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points')
plot.set_xlabel("Client Types")
plot.set_ylabel("Percentage distribution of Client Types")
plot.set_title("Percentages of Different Client Types")


# TARGET VS Type of Clients

client_target = pd.pivot_table(data = updated_curr_app, columns = 'TARGET' ,index = 'NAME_CLIENT_TYPE' , values = 'SK_ID_CURR', aggfunc = 'count')
client_target['Total'] = client_target[0] + client_target[1]
client_target = (client_target.div(client_target.Total, axis = 0)*100).drop('Total', axis = 1)


# Plotting bar plot

plt2 = client_target.plot.bar(ax = ax[1], figsize = (15,7))
for bar in plt2.patches:
        plt2.annotate("{}%".format(format(bar.get_height(), '.2f')), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points')

        
plot.set_xlabel("Client Types")
plot.set_ylabel("Percentage distribution of Client Types")
plot.set_title("Percentages of Different Client Types depends on TARGET")

plt2.legend(bbox_to_anchor = (1.0,1.0))
plt.show()
```

`Inference`: We can see here that nearly 52% of clients in current applications are 'Repeater'. Only 39% of New Applicants applied for the loan

- It is obvious that maximum clients of TARGET = 1, falls under Client Type 'New'
- So, banks/similar institutions has to take care the most if the client is new for loan application process

## Finding relation between difference of Amount Applied and Amount Credited with Amount Goods Price Vs Approved and Canceled Loans

```python
app_canc_loans = prev_app[(prev_app.NAME_CONTRACT_STATUS == 'Approved') | (prev_app.NAME_CONTRACT_STATUS == 'Canceled')]
app_canc_loans['APP_CREDIT_DIFF'] = app_canc_loans.AMT_CREDIT - app_canc_loans.AMT_APPLICATION
app_canc_loans['GOODS_CREDIT_DIFF'] = app_canc_loans.AMT_CREDIT - app_canc_loans.AMT_GOODS_PRICE
app_canc_loans['APP_GOODS_DIFF'] = abs(app_canc_loans.AMT_APPLICATION - app_canc_loans.AMT_GOODS_PRICE)

app_canc_loans
```

### Binning of Amount Differences into Categories

```python
app_canc_loans.APP_GOODS_DIFF.describe()

## We can infer here that there is no difference betwwen amount of Goods mortgaged and Amount on Application.
### So, it is obviuos that Clients are applying for loan on the goods with 100% of its true value in the application.
```

```python
app_canc_loans['GOODS_CREDIT_DIFF'].min()
```

```python
app_canc_loans['APP_CREDIT_BIN'] = pd.cut(x= app_canc_loans.APP_CREDIT_DIFF, bins = [-3500000,-2500000, -1000000, 500000, 1000000, 2000000], labels = ['very-low', 'low', 'moderate', 'high', 'very-high'])
app_canc_loans['GOODS_CREDIT_BIN'] = pd.cut(x= app_canc_loans.GOODS_CREDIT_DIFF, bins = [-3000000,-2000000, -1000000,0, 500000, 1000000], labels = ['very-low', 'low', 'moderate', 'high', 'very-high'])

app_canc_loans.head()
```

```python
app_credit_pivot = pd.pivot_table(data = app_canc_loans , columns = 'NAME_CONTRACT_STATUS', index = 'APP_CREDIT_BIN', values = 'SK_ID_CURR', aggfunc = 'count')
app_credit_pivot['Total'] = app_credit_pivot.sum(axis = 1)

plt.figure(figsize = (12,5))
sns.heatmap((app_credit_pivot.div(app_credit_pivot.Total, axis = 0)*100).drop('Total', axis = 1).fillna(0), fmt = 'g', cmap = 'Blues', annot = True, center = 50)

plt.show()
```

`Inference` : It is interseting to see that there are loans where the companies credited the amount greater than the amount of application. 

- It is seen that for those loans where the loan amount credited is higher than the loan amount credited, there are almost flawless approval of loans i.e, near to 100%

- Also, if the difference between loan credited and loan applied is very low, there too we infer 100% approval of loans

```python
good_credit_pivot = pd.pivot_table(data = app_canc_loans , columns = 'NAME_CONTRACT_STATUS', index = 'GOODS_CREDIT_BIN', values = 'SK_ID_CURR', aggfunc = 'count')
good_credit_pivot['Total'] = good_credit_pivot.sum(axis = 1)

plt.figure(figsize = (12,6))
sns.heatmap((good_credit_pivot.div(good_credit_pivot.Total, axis = 0)*100).drop('Total', axis = 1).fillna(0), cmap = 'Blues', fmt = 'g', annot = True, center = 49.999)
plt.show()
```

`Inference`: It has high contrast with the insights of differnce between Amount Application and Amount Credited Vs Approval Status plot which is obtained above.

- If the differnce between the Amount of goods mortgaged and Amount of Credit is very-high and ver-low, there are only 60% positive response i.e, Approved Status of Applications

- Applications with difference amount in between -20Lakhs to +5Lakhs has about 98% of Approved Loans

## Finding the successful Chanel Type

```python
channel_status = pd.pivot_table(data = prev_app, columns = 'NAME_CONTRACT_STATUS', index = 'CHANNEL_TYPE', values = 'SK_ID_CURR', aggfunc = 'count').fillna(0)
total_application = len(prev_app)
channel_status = (channel_status.div(total_application)*100)


# Plotting Heatmap showing Percent of Channel Type
plt.figure(figsize = (12,6))
sns.heatmap(channel_status, cmap = 'BuPu', annot = True, fmt = 'g', center = 3.125)
plt.show()
```

`Inference`:
- The most familiar channel type for Approved loans is 'Country-wide'
- The least familiar channel type is 'Car Dealer'
